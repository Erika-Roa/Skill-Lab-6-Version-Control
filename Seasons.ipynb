```python
#imports necessary modules from IPython for displaying images

from IPython.display import display, Image
import os

class Seasons:
#constructor to initialize the season class
    def __init__(self, month=1, day=1):
        self.month = month
        self.day = day
        self.season = None
        self.image_path = None

#getters and setters
    def get_month(self):
        return self.month

    def set_month(self, month):
        self.month = month

    def get_day(self):
        return self.day

    def set_day(self, day):
        self.day = day
#gets the season
    def get_season(self):
        if self.month in [1, 2, 3]:
            self.season = "winter"
        elif self.month in [4, 5, 6]:
            self.season = "spring"
        elif self.month in [7, 8, 9]:
            self.season = "summer"
        elif self.month in [10, 11, 12]:
            self.season = "fall"
        else:
            return "Invalid date! Try again"
        
        if self.month % 3 == 0 and 21 <= self.day <= 31:
            season_changes = {"winter": "spring", "spring": "summer", "summer": "fall", "fall": "winter"}
            self.season = season_changes.get(self.season, self.season)
        
        return self.season
#gets and sets image path
    def get_image_path(self):
        return self.image_path
    
    def set_image_path(self, path):
        self.image_path = path
```
